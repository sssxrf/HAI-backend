```python
import pandas as pd
import plotly.express as px
import os
```

```python
sheet = "/Users/daveyonkers/Downloads/User Evaluation (Responses) - Form Responses 1.csv"

df = pd.read_csv(sheet)
df.head()
```

```python
for i in range(len(df.columns)):
    
    # if not integer data, then we can't plot it
    if df.dtypes[i] != "int64":
        continue
        
    question = df.columns[i]
    responses = df[question]
    
    # sum up the ratings and put it into response_df
    options = ["Disagree", "Slightly Disagree", "Indifferent", "Slightly Agree", "Agree"]
    response_df = pd.DataFrame({"Rating": options, "Count": [0 for _ in options]})
    for i in range(len(response_df)):
        response_df.loc[i, "Count"] = sum(df[question] == i+1)

    # plot it yuh
    # did not include title, because it is too long and doesn't wrap to multiple lines
    font_color = "rgb(78, 84, 112)"
    fig = px.bar(response_df, x="Rating", y="Count", color_discrete_sequence=['rgb(181, 215, 228)'], template="simple_white")
    fig.update_yaxes(showgrid=True, linecolor=font_color)
    fig.update_xaxes(linecolor=font_color)
    fig.update_layout(
        font_color=font_color,
        font_family="Nunito",  # matches presentation font
    )  
    
    # only if you want to print
#     fig.show()
    
    # create a directory to save the plots, if it does not already exist
    plotting_dir = "plots/"
    if not os.path.exists(plotting_dir):
        os.mkdir(plotting_dir)
        
    # save image
    image_path = plotting_dir + question
    fig.write_image(image_path + ".png")
    fig.write_image(image_path + ".pdf")
    
            
```

```python
question = df.columns[-1]

options = ["None", "Beginner", "Expert", "Fluent"]
response_df = pd.DataFrame({"Rating": options, "Count": [0 for _ in options]})
for i in range(len(response_df)):
    response_df.loc[i, "Count"] = sum(df[question] == options[i])

fig = px.bar(response_df, x="Rating", y="Count", color_discrete_sequence=['rgb(181, 215, 228)'], template="simple_white")
fig.update_yaxes(showgrid=True, linecolor=font_color)
fig.update_xaxes(linecolor=font_color)
fig.update_layout(
    font_color=font_color,
    font_family="Nunito",  # matches presentation font
) 

# save image
image_path = plotting_dir + question
fig.write_image(image_path + ".png")
fig.write_image(image_path + ".pdf")
```
